#Café Orders & Inventory Tracker
This project simulates a small café system that handles daily orders. The program stores menu items with prices, keeps track of inventory, processes each order using loops, calculates bills, updates the stock, and creates a summary at the end of the day. It helps show how lists and dictionaries work together in real-world situations while using only basic Python



```python
# Café Order Management System
#Menu & prices
menu = {
    "coffee": 250,
    "cappuccino": 450,
    "iced coffee": 300,
    "black coffee": 200,
    "tea": 150,
    "green tea": 150,
    "hot chocolate": 400,
    "sandwich": 300,
    "cake slice": 250,
    "pastri": 300,
    "donut": 250,
    "brownie": 180,
    "juice": 150,
    "mineral water": 90
}
#Make  Inventory for item
inventory = {
    "coffee": 95,
    "cappuccino": 75,
    "iced coffee": 25,
    "black coffee": 20,
    "tea": 80,
    "green tea": 50,
    "hot chocolate": 40,
    "sandwich": 10,
    "cake slice": 38,
    "pastri": 30,
    "donut": 25,
    "brownie": 18,
    "juice": 100,
    "mineral water": 200
}

# To store daily summary
daily_sales = {}
total_revenue = 0

print("Welcome to the Café!\n")
print("Today's Menu:")
for item, price in menu.items():
    print(f"- {item.title()}: Rs. {price}")
print("\n-----------------------------\n")

# Process unlimited orders until user stops
while True:
    print("\nEnter a customer's order:")

    order_item = input("Item name (or 'done' to finish day): ").lower()

    # End-of-day summary trigger
    if order_item == "done":
        break

    # Check if item exists
    if order_item not in menu:
        print("Item not found in menu. Try again.")
        continue

    # Quantity
    try:
        quantity = int(input("Quantity: "))
    except ValueError:
        print("Invalid quantity. Enter a number.")
        continue

    # Check stock
    if quantity > inventory[order_item]:
        print(f"Sorry, only {inventory[order_item]} left in stock.")
        continue

    # Calculate bill
    price = menu[order_item]
    cost = price * quantity
    print(f"Added to bill: Rs. {cost}")

    # Update inventory
    inventory[order_item] -= quantity

    # Update daily sales
    if order_item in daily_sales:
        daily_sales[order_item] += quantity
    else:
        daily_sales[order_item] = quantity

    # Add to revenue
    total_revenue += cost

# End of Day Summary

print("\n==============================")
print("END OF DAY SUMMARY")
print("==============================")

print("\nSales Summary:")
for item, qty in daily_sales.items():
    print(f"- {item.title()} sold: {qty}")

print("\nRemaining Inventory:")
for item, stock in inventory.items():
    print(f"- {item.title()}: {stock} left")

print(f"\n Total Revenue: Rs. {total_revenue}")

print("\nThank you! Café system closed.")
```
